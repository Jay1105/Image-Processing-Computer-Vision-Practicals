```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/wrld1.webp")
cv2_imshow(img)
```

```python
img2 = cv2.imread("/content/wrld2.jfif")
cv2_imshow(img2)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img2)
```

```python
print(img.shape)

cropped_img1 = img[100:300, 200:500]
cv2_imshow(cropped_img1)
```

```python
print(img2.shape)

cropped_img2 = img2[50:250, 400:700]
cv2_imshow(cropped_img2)
```

```python
add_img = cv2.add(cropped_img1, cropped_img2)
subtract_img = cv2.subtract(cropped_img1, cropped_img2)

plt.subplot(2, 2, 1)
plt.title("A")
plt.imshow(cropped_img1)

plt.subplot(2, 2, 2)
plt.title("B")
plt.imshow(cropped_img2)

plt.subplot(2, 2, 3)
plt.title("Add")
plt.imshow(add_img)

plt.subplot(2, 2, 4)
plt.title("Subtract")
plt.imshow(subtract_img)
```

```python
and_img = cv2.bitwise_and(cropped_img1, cropped_img2, mask = None)
or_img = cv2.bitwise_or(cropped_img1, cropped_img2, mask = None)
not_img = cv2.bitwise_not(cropped_img2, mask = None)
xor_img = cv2.bitwise_xor(cropped_img1, cropped_img2, mask = None)

plt.subplot(3, 2, 1)
plt.title("A")
plt.imshow(cropped_img1)

plt.subplot(3, 2, 2)
plt.title("B")
plt.imshow(cropped_img2)

plt.subplot(3, 2, 3)
plt.title("And")
plt.imshow(and_img)

plt.subplot(3, 2, 4)
plt.title("Or")
plt.imshow(or_img)

plt.subplot(3, 2, 5)
plt.title("Not")
plt.imshow(not_img)

plt.subplot(3, 2, 6)
plt.title("Xor")
plt.imshow(xor_img)
```
